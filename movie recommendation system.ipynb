```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
dataset=pd.read_csv('ratings.csv')
```

```python
dataset.head()
```

```python
movie_titles=pd.read_csv('movies.csv')
```

```python
movie_titles.head()
```

```python
dataset=pd.merge(dataset,movie_titles,on='movieId')
dataset.head()
```

```python
dataset.describe()
```

```python
ratings=pd.DataFrame(dataset.groupby('title')['rating'].mean())
```

```python
ratings.head()
```

```python
ratings['no_of_ratings']=dataset.groupby('title')['rating'].count()
ratings.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ratings['rating'].hist(bins=50)
```

```python
ratings['no_of_ratings'].hist(bins=60)
```

```python
import seaborn as sns
sns.jointplot(x='rating',y='no_of_ratings',data=ratings)
```

```python
movie_matrix=dataset.pivot_table(index='userId',columns='title',values='rating')
```

```python
movie_matrix.head()
```

```python
ratings.sort_values('no_of_ratings',ascending=False).head(10)
```

```python
fg_rating=movie_matrix['Forrest Gump (1994)']
sr_rating=movie_matrix['Shawshank Redemption, The (1994)']
```

```python
similar_to_fg=movie_matrix.corrwith(fg_rating)
```

```python
similar_to_sr=movie_matrix.corrwith(sr_rating)
```

```python
similar_to_fg.head()
```

```python
similar_to_sr.head()
```

```python
corr_fg = pd.DataFrame(similar_to_fg, columns=['Correlation'])
corr_fg.dropna(inplace=True)
corr_fg.head()
corr_sr = pd.DataFrame(similar_to_sr, columns=['correlation'])
corr_sr.dropna(inplace=True)
corr_sr.head()
```

```python
corr_fg=corr_fg.join(ratings['no_of_ratings'])
corr_sr=corr_sr.join(ratings['no_of_ratings'])
```

```python
corr_fg[corr_fg['no_of_ratings']>100].sort_values(by='Correlation',ascending=False).head(10)
```

```python
corr_sr[corr_sr['no_of_ratings']>100].sort_values(by='correlation',ascending=False).head(10)
```
